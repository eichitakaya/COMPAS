# めあて：dicom画像を扱おう

### DICOMデータの中身を見る

pydicomというライブラリでdicomデータを扱う．<br> インストールされていない場合は，pip install pydicom

```python
import pydicom
```

```python
# dicomデータの読み込み
dcm = pydicom.read_file("dataset/negative/axial/1.2.392.200036.9142.10002202.1020330992.3.20190507181142.96409/DicomFile.dcm")
```

```python
# dicom画像の本体（ピクセルの行列）はpixel_arrayで表示
dcm_img = dcm.pixel_array
dcm_img
```

```python
# 実際に描画してみる
from PIL import Image
import numpy as np
Image.fromarray(np.uint8(dcm_img))
```

テキトーに8ビットに変換してしまうと，上記のようになってしまうので，適切なウィンドウ処理を施す（参考記事は[こちら](https://guarana001.hateblo.jp/entry/2020/07/26/134446)） <br> 変換の式は以下の通り．


<img src="window.png">

ただし，nは変換後のbit数，xは画素値, wcはwindow center, wwはwindow width

```python
# プログラムに落とし込むとこんな感じ
dcm_wc = dcm.WindowCenter
dcm_ww = dcm.WindowWidth

window_max = dcm_wc + dcm_ww/2                     
window_min = dcm_wc - dcm_ww/2
dcm_img = 255*(dcm_img - window_min)/(window_max - window_min)    
dcm_img[dcm_img > 255] = 255
dcm_img[dcm_img < 0] = 0 
```

```python
Image.fromarray(np.uint8(dcm_img))
```
